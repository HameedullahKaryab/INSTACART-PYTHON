## If- Statements with User-Defined Functions

```python
# Create a subset of 1m rows
df = df_ords_prods_merged[:1000000]
```

```python
df.shape
```

```python
# Define a function

def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif (row['prices'] > 15):
        return 'High range product'
    else: return 'Not enough data'
        
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Max function
df['prices'].max()
```

# Q4. Creating a new column called (busiest_period_of_day)

```python
# First lets Check the frequency of values in the "order_hour_of_day" column 

df_ords_prods_merged['order_time_of_day'].value_counts()
```

####  The senior technical officer at Instacart wants these to be grouped as "Most orders" = 10, 11,14,15,13,12,16,9 Average orders = 17,8,18,19,20,7,21,22  Fewest orders = 23,6,0,1,5,2,4,3

```python
# If-Statements with For-loop

hour_of_day = []

for value in df_ords_prods_merged["order_time_of_day"]:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        hour_of_day.append("Most orders")
    elif value in [17, 8, 18, 19, 20, 7, 21, 22]:
        hour_of_day.append("Average orders")
    else:
        hour_of_day.append("Fewest orders")
```

```python
# Add result to the dataframe

df_ords_prods_merged['busiest_period_of_day'] = hour_of_day
```
